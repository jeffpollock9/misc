$$
\alpha = \int_0^1 x dx
$$

```r
runs <- 100000
xs <- runif(runs, -0.5, 0.5)
ys <- runif(runs, -0.5, 0.5)
in_circle <- xs^2 + ys^2 <= 0.5^2
mc_pi <- sum(in_circle) / runs * 4
plot(xs, ys, pch='.', col=ifelse(in_circle, "blue", "grey"), xlab='', ylab = '', asp=1,
     main=paste("MC Approximation of pi =", mc_pi))
```
